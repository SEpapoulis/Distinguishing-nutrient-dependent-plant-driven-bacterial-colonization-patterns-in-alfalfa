# Annotation of ASVs Using Sanger Amplicons

We will first align all SynCom sanger amplicons to one another to find a threshold for the percent identity, then use that identity to annotate ASVs.

 Required dependencies to reproduce Notebook results:
    - Nucleotide-Nucleotide BLAST 2.7.1+
